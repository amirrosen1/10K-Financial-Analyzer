```python
from bs4 import BeautifulSoup

def extract_tables_from_html(html_content):
    """
    Extracts all tables from HTML content and returns them as plain text.
    """
    soup = BeautifulSoup(html_content, "html.parser")
    tables = soup.find_all("table")

    extracted_tables = []
    for idx, table in enumerate(tables):
        rows = table.find_all("tr")
        table_data = []
        for row in rows:
            cells = row.find_all(["th", "td"])
            row_data = [cell.get_text(strip=True) for cell in cells]
            table_data.append(row_data)
        extracted_tables.append(table_data)
    return extracted_tables

# Example: Load your HTML content
with open("part_8_financial_statements.html", "r", encoding="utf-8") as file:
    html_content = file.read()

tables = extract_tables_from_html(html_content)

# Print one table as plain text for inspection
for row in tables[2]:  # Process the first table
    print("\t".join(row))
```

```python
# def format_table_for_prompt(table):
#     """
#     Formats a table into plain text for use in an LLM prompt.
#     """
#     formatted_table = "\n".join(["\t".join(row) for row in table if any(row)])
#     return formatted_table

# # Example: Format the first table
# formatted_table = format_table_for_prompt(tables[0])
# print("Formatted Table:")
# print(formatted_table)
def format_table_for_prompt(table):
    """
    Cleans and formats a table into plain text with proper alignment for use in an LLM prompt.
    """
    # Remove empty rows and normalize the table
    table = [row for row in table if any(cell.strip() for cell in row)]

    # Find the maximum number of columns in the table
    max_columns = max(len(row) for row in table)

    # Normalize all rows to have the same number of columns
    normalized_table = [row + [""] * (max_columns - len(row)) for row in table]

    # Determine the maximum width of each column
    column_widths = [max(len(cell.strip()) for cell in col) for col in zip(*normalized_table)]

    # Format each row with consistent column widths
    formatted_rows = []
    for row in normalized_table:
        formatted_row = "  ".join(f"{cell.strip():<{column_widths[i]}}" for i, cell in enumerate(row))
        formatted_rows.append(formatted_row)

    # Join all rows into a single formatted table string
    formatted_table = "\n".join(formatted_rows)
    return formatted_table

# Example: Format the first table
formatted_table = format_table_for_prompt(tables[2])
print("Formatted Table:")
print(formatted_table)

```

```python
# Create a structured prompt using the formatted table
prompt = f"""
Below is a table from a company's financial statements:

{formatted_table}

Question: What is the profit for the most recent year in the table above?
"""

print("Prepared Prompt:")
print(prompt)
```
